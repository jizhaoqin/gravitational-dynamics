```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rotateFrame(pos, angle):
    """rotate a 2D shape 'pos' by 'angle'

    Args:
        pos (nparray): position list
        angle (nparray): angle list

    Returns:
        (nparray): rotated pos list
    """
    # rotation matrices
    rotation_matrix = np.array([[+np.cos(angle), +np.sin(angle)],
                                [-np.sin(angle), +np.cos(angle)]]).transpose((2, 0, 1))

    rotated_pos = np.zeros_like(pos)
    rotated_pos[:, 0] = np.sum(rotation_matrix[:, 0]*pos, axis=-1)
    rotated_pos[:, 1] = np.sum(rotation_matrix[:, 1]*pos, axis=-1)
    return rotated_pos
```
