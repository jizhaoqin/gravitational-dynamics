```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
```

```python
KDKEllipse = np.loadtxt("./KDK-ellipse-1e-1.txt", skiprows=5)
DKDEllipse = np.loadtxt("./DKD-ellipse-1e-1.txt", skiprows=5)
EulerEllipse = np.loadtxt("./Euler-ellipse-v-first-1e-1.txt")
DKDAdaptiveEllipse = np.loadtxt("./adaptive-DKD-stepsize.txt")
DKDAdaptiveEllipseLargeT = np.loadtxt("./adaptive-DKD-large-t.txt")
KDKAdaptiveEllipse = np.loadtxt("./adaptive-KDK-stepsize.txt")
length = len(KDKEllipse)
```

```python
# energy plot
# plt.scatter(EulerEllipse[:length, 0], EulerEllipse[:length, -1], s=1, label="Euler: update vel first")
# plt.plot(KDKEllipse[:, 0], KDKEllipse[:, -1], alpha=1, label="KDK")
plt.plot(KDKAdaptiveEllipse[:, 0], KDKAdaptiveEllipse[:, -1], alpha=1, label="KDK adaptive")
plt.plot(DKDEllipse[:, 0], DKDEllipse[:, -1], alpha=1, label="DKD")
plt.plot(DKDAdaptiveEllipse[:, 0], DKDAdaptiveEllipse[:, -1], alpha=1, label="DKD adaptive")
plt.title("Total Energy of Earth, dt=1e-2 year")
plt.xlabel("time (year)")
plt.ylabel("total energy (M_sun * AU^2 / year^2)")
plt.legend()
plt.grid()
# plt.savefig("Energy.png", dpi=300)
plt.show()
```
